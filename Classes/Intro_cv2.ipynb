Una vez se confirmó las librerías descargadas, se usará la imagen de Lenna y opencv.

```python
Lenna = cv2.imread('Lenna.jpg')

plt.imshow(Lenna)
plt.show()
```

La imagen no salió con los colores reales por el mapa de color que usa opencv. Al usar la función cv2.imread la imagen cambia la gama de colores que después se imprimirán.

OpenCV usa al padrón BGR. El padrón actual en toda foto el RGB.

Para que cambie de BGR a RGB se usa el siguiente comando:

```python
Lenna = cv2.imread('Lenna.jpg')

# 1. Usando opencv
# Lenna_RGB = cv2.cvtColor(Lenna,cv2.COLOR_BGR2RGB)
# plt.imshow(Lenna_RGB)

# 2. Usando funciones de python

# - Se saca por canal
print('Dimensión de la imagen: ', Lenna.shape)
# - Tiene 3 canales para BGR
blue = Lenna[:,:,0]
green = Lenna[:,:,1]
red = Lenna[:,:,2]

# Comprobando los colores escalando a grises
# plt.imshow(blue,cmap='gray')
# plt.imshow(green,cmap='gray') # Es el que más se parece
# plt.imshow(red,cmap='gray')

imagen_RGB = cv2.merge((red, green, blue))
plt.imshow(imagen_RGB)
plt.show()

imagen_gray = cv2.cvtColor(Lenna,cv2.COLOR_BGR2GRAY)
plt.imshow(imagen_gray,'gray')
plt.show()
```

Se procede a recortar la imagen tomando una parte de la matriz de la imagen
